```python
plt.figure(figsize=(10,6))
sns.histplot(df['Fare'],bins=20,kde=True,)
plt.title("Fare Distribution")
plt.xlabel("Fare")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['Age'],bins=30,edgecolor='k',alpha=0.7)
plt.title("Distribution Of Ticket")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(10,6))
df['Pclass'].value_counts().plot(kind='bar',color=['Blue','Red'])
plt.title("Distribution of Pclass")
plt.xlabel("Pclass")
plt.ylabel("Frequency")
plt.show()
```

```python
df['Sex'].value_counts().plot.bar()
```

```python
plt.figure(figsize=(10,6))
df['Sex'].value_counts().plot(kind='bar',color=['Blue','Red'])
plt.title("Distribution of Gender")
plt.xlabel("Sex")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['Fare'],df['Parch'],alpha=0.5)
plt.title("Fare Vs Parch")
plt.xlabel("Fare")
plt.ylabel("Parch")
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['Survived'],df['Pclass'],alpha=0.7)
plt.title("Survived Vs Pclass")
plt.xlabel("Survived")
plt.ylabel("Pclass")
plt.show()
```

```python

correlation_matrix = df[['PassengerId','Survived','Pclass','Age','SibSp','Parch','Fare']].corr()
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
